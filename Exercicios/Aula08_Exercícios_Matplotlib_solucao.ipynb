## **Exercício 2 – Plotagem Básica**

**a) Escreva um código em Python que plote um gráfico de linha simples mostrando os valores de crescimento de uma planta nos 5 primeiros dias: `dias = [1,2,3,4,5]`, `altura = [2, 3, 4, 5, 7]`.**

**b) Adicione títulos e rótulos apropriados aos eixos x e y.**

````python
import matplotlib.pyplot as plt

# Dados
dias = [1, 2, 3, 4, 5]
altura = [2, 3, 4, 5, 7]

# Gráfico de linha simples
plt.plot(dias, altura)
plt.xlabel('Dias')
plt.ylabel('Altura (cm)')
plt.title('Crescimento da Planta nos 5 Primeiros Dias')

# Use sempre plt.show() no final para exibir o gráfico, especialmente fora de notebooks.
plt.show()
````

**a) Escreva uma linha de código usando pandas para gerar um gráfico de barras das vendas por produto.**

````python
#OPÇÃO 1: VERTICAL
import pandas as pd
df = pd.DataFrame({
    "produto": ["A", "B", "C"],
    "vendas": [150, 250, 180]
})

df.plot.bar(x='produto', y='vendas')
plt.ylabel("Vendas")
plt.xlabel("Produto")
plt.title("Vendas por Produto")
plt.show()
````

````python
# OPÇÃO 2: HORIZONTAL
import pandas as pd
df = pd.DataFrame({
    "produto": ["A", "B", "C"],
    "vendas": [150, 250, 180]
})

df.plot.barh(x='produto', y='vendas')
plt.ylabel("Vendas")
plt.xlabel("Produto")
plt.title("Vendas por Produto")
plt.show()
````

## **Exercício 6 – Subplots e Personalização**

Escreva um código para criar uma figura com **duas subplots lado a lado**:

- A primeira subplot deve ser um gráfico de linha de x=[1,2,3] e y=[2,4,3].
- A segunda subplot um gráfico de barras para as categorias ['a','b','c'] com valores [4,6,5].Dê títulos a cada subplot e um título geral para a figura.

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(10,4))  # 1 linha, 2 colunas

# Primeira subplot - linha
x = [1, 2, 3]
y = [2, 4, 3]
axs[0].plot(x, y, marker='o')
axs[0].set_title("Gráfico de Linha")

# Segunda subplot - barras
categorias = ['a', 'b', 'c']
valores = [4, 6, 5]
axs[1].bar(categorias, valores)
axs[1].set_title("Gráfico de Barras")

fig.suptitle("Exemplo de Subplots em Matplotlib")
# Dica: plt.tight_layout() ajuda os gráficos a não ficarem sobrepostos.
plt.tight_layout()
plt.show()
````
