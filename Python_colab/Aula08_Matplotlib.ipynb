# API Básica: Figuras e Subplotagens

```python
fig, ax = plt.subplots()  # Cria uma figure e um eixo (axes)
# A figure é como uma "tela" ou folha onde os gráficos serão desenhados.
# O eixo (ax) representa a área dentro da figura onde o gráfico realmente aparece (com os eixos X e Y).

ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='green', linewidth=3, marker='s', markersize=6)  # Plota uma linha simples
ax.plot([1, 2, 3, 4], [15, 25, 30, 20], color='orange', linewidth=3, marker='o', markersize=6)
# Eixo X: 1, 2, 3, 4
# Eixo Y: 10, 20, 25, 30
# A função plot() pertence ao objeto ax, que permite controle detalhado do gráfico (como título, rótulos, legendas etc.).
ax.set_xticks([1, 2, 3, 4])  # Define os valores do eixo X
ax.set_yticks([10, 15, 20, 25, 30])  # Define os valores do eixo Y

plt.show()
# Exibe a figura final na tela.
# Sem esse comando, o gráfico pode não aparecer
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
# Cria uma grade de subplots com 2 linhas e 2 colunas — totalizando 4 gráficos.
# fig é a figura principal (como uma folha grande).
# axes é um array 2x2 com os 4 eixos (um para cada gráfico).
# figsize=(8, 6) define o tamanho total da figura (largura 8, altura 6, em polegadas).

for i in range(2):
    for j in range(2):
        axes[i, j].plot([1,2,3], [j+1, (i+1)*2, (i+j)*3])
        axes[i, j].plot([1,2,3], [j+1, (i+1)*3, (i+j)*2])
        # Cria um gráfico de linha nos eixos correspondentes à posição [i, j].
        # Os valores de Y mudam dinamicamente com base nos índices i e j, para que cada gráfico tenha dados diferentes.

        axes[i, j].set_title(f'Plot {i},{j}')
        # Define um título individual para cada subplot, com os índices atuais (linha, coluna), como "Plot 0,1".

plt.tight_layout()
# Ajusta o espaçamento automático entre os subplots.
# Evita que títulos, rótulos ou gráficos fiquem sobrepostos.

plt.show()
```

```python
fig, ax = plt.subplots()
# Cria uma figura (fig) e um eixo (ax) para o gráfico.
# Isso é a base para desenhar.

ax.plot([1, 2, 3, 4], [2, 4, 8, 16], 'g--o', label='Crescimento')  # Linha verde tracejada com círculos
# [1, 2, 3, 4]: valores no eixo X
# [2, 4, 8, 16]: valores no eixo Y
# g--o': estilo da linha:
#   g: cor verde (green)
#   --: linha tracejada
#   o: círculos nos pontos de dados
# label='Crescimento': Define o nome da curva para aparecer na legenda.

ax.legend()
# Exibe a legenda no gráfico com o texto "Crescimento".

plt.show()
```

```python
# Extra: trazendo um dado comparativo ao gráfico anterior
import matplotlib.pyplot as plt

# Cria a figura e os eixos
fig, ax = plt.subplots()

# Primeira curva: crescimento exponencial
ax.plot([1, 2, 3, 4], [2, 4, 8, 16], 'g--o', label='Crescimento')  # Linha verde tracejada com círculos

# Segunda curva: crescimento linear
ax.plot([1, 2, 3, 4], [2, 4, 6, 8], 'b-^', label='Linear')  # Linha azul sólida com triângulos

# Adiciona a legenda para identificar cada linha
ax.legend()

# Exibe o gráfico
plt.show()
```

```python
fig, ax = plt.subplots()
# Cria uma figura (fig) e um eixo (ax) para o gráfico.

# dados a serem plotados.
x = [0, 1, 2, 3]
y = [10, 3, 8, 7]

ax.plot(x, y, color='#FF5733', linestyle='-.', marker='s', label="Exemplo")
# color='#FF5733': define a cor da linha com um código hexadecimal (um tom laranja-avermelhado).
# linestyle='-.': usa uma linha traço-ponto.
# marker='s': usa quadrados (square) nos pontos.
# label='Exemplo': define o nome para a legenda.

ax.set(title='Título do Gráfico', xlabel='Eixo X', ylabel='Eixo Y')
# Define o título do gráfico e os nomes dos eixos.

ax.set_xlim(-1, 4)
ax.set_ylim(0, 12)
# Define os limites visíveis no gráfico:
# Eixo X de -1 até 4
# Eixo Y de 0 até 12

ax.grid(True)
# Ativa a grade (linhas de referência) no gráfico para facilitar a leitura dos valores.

ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot([1,2,3], [10, 20, 15])
# Plota uma linha ligando os pontos:
# (1, 10) → (2, 20) → (3, 15)

ax.text(2, 20, "Pico!", fontsize=12, color='red')
# Adiciona um texto fixo no gráfico.
# (2, 20): posição do texto (coordenadas X e Y).
# "Pico!": o texto que será exibido.
# fontsize=12: tamanho da fonte.
# color='red': cor do texto.

ax.annotate('Mínimo', xy=(1, 10), xytext=(1, 13),
            arrowprops=dict(facecolor='blue', shrink=0.05))
# Adiciona uma anotação com seta, apontando para um ponto específico:
# 'Mínimo': o texto da anotação.
# xy=(1, 10): ponto alvo, onde a seta vai apontar.
# xytext=(1, 13): posição do texto, um pouco acima.
# arrowprops=...: define o estilo da seta:
# facecolor='blue': seta azul.
# shrink=0.05: encurta um pouco a seta para deixar mais elegante.

plt.show()
```

```python
from matplotlib.patches import Rectangle, Circle
# Importa as classes para criar formas geométricas: retângulo e círculo.
# Elas são chamadas de "patches", e podem ser inseridas nos eixos (ax).

fig, ax = plt.subplots()
ax.plot([0,1,2], [0,1,0])
# Plota uma curva simples em forma de "V" invertido, com os pontos:
# (0,0) → (1,1) → (2,0)

rect = Rectangle((0.5, 0), 0.8, 1, color='orange', alpha=0.3)
# Retângulo:
# (0.5, 0): posição inferior esquerda do retângulo (x, y).
# 0.8: largura.
# 1: altura.
# color='orange': cor laranja.
# alpha=0.3: transparência de 30% (valores entre 0 e 1).

circ = Circle((1.5, 0.5), 0.2, color='green', fill=True)
# Circulo
# (1.5, 0.5): centro do círculo (x, y).
# 0.2: raio.
# color='green': cor verde.
# fill=True: preenche o interior (padrão).

ax.add_patch(rect) # adiciona a forma retângulo ao gráfico.
ax.add_patch(circ) # adiciona a forma círculo ao gráfico.
plt.show()
```

# Tipos de Gráficos Comuns – Exemplos

```python
x = [2018, 2019, 2020, 2021]
y = [100, 120, 90, 130]

plt.plot(x, y, marker='o')
# marker='o': adiciona um círculo em cada ponto do gráfico para facilitar a visualização.

plt.title("Vendas ao longo do tempo")
plt.xlabel("Ano")
plt.ylabel("Quantidade")
plt.show()
```

```python
categorias = ['A', 'B', 'C']
# Define os rótulos do eixo X, representando cada grupo ou categoria.

valores = [23, 45, 12]
# Define os valores do eixo Y, ou seja, a altura de cada barra.

plt.bar(categorias, valores)
# Cria um gráfico de barras verticais.
# Cada categoria terá uma barra com altura proporcional ao seu valor.

plt.xlabel("Categoria")
plt.ylabel("Frequência")
plt.title("Distribuição por categoria")
plt.show()
```

```python
# Extra 1: versão horizontal do gráfico anterior
import matplotlib.pyplot as plt

categorias = ['A', 'B', 'C']
# Rótulos do eixo Y agora, pois o gráfico será horizontal.

valores = [23, 45, 12]
# Comprimento das barras na horizontal.

plt.barh(categorias, valores)
# Cria um gráfico de barras horizontais.

plt.xlabel("Frequência")
plt.ylabel("Categoria")
plt.title("Distribuição por categoria (horizontal)")
plt.show()
```

```python
# Extra 2: Versão anterior com dados comparativos

import matplotlib.pyplot as plt
import numpy as np

# Dados
categorias = ['A', 'B', 'C']
valores_2024 = [23, 45, 12]
valores_2025 = [30, 38, 18]

# Índices das categorias no eixo Y
y = np.arange(len(categorias))

# Largura da barra
altura = 0.4

# Criação do gráfico de barras horizontais lado a lado
plt.barh(y - altura/2, valores_2024, height=altura, label='2024', color='skyblue')
plt.barh(y + altura/2, valores_2025, height=altura, label='2025', color='lightgreen')

# Rótulos e título
plt.xlabel("Frequência")
plt.ylabel("Categoria")
plt.yticks(y, categorias)
plt.title("Distribuição por categoria (comparativo)")
plt.legend()

# Mostrar o gráfico
plt.show()
```

```python
labels = ['Sim', 'Não', 'Talvez']
sizes = [50, 30, 20]
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
# Argumentos:
# sizes: tamanhos relativos de cada fatia.
# labels: nomes exibidos em cada fatia.
# autopct="%1.1f%%": exibe porcentagens dentro das fatias com 1 casa decimal (ex: 50.0%).

plt.title("Preferências")
plt.show()
```

```python
import numpy as np
x = np.random.randn(100)
# Gera 100 valores aleatórios de uma distribuição normal (gaussiana) com média 0 e desvio padrão 1.

y = 2 * x + np.random.rand(100)
# Cria os valores do eixo Y com uma relação linear aproximada com x.
# 2 * x: cria uma relação linear (declive 2).
# + np.random.rand(100): adiciona um pequeno ruído aleatório (entre 0 e 1) para tornar os dados mais realistas.

plt.scatter(x, y)
# Cria um gráfico de dispersão com os valores de x e y.
# Cada ponto representa uma observação (x, y).
# Ideal para analisar correlação e tendências visuais.

plt.title("Relação entre duas variáveis")
plt.xlabel("Variável X")
plt.ylabel("Variável Y")
plt.show()
```

```python
# Extra: gráfico anterior com dados comparativos
import matplotlib.pyplot as plt
import numpy as np

# Conjunto 1
x1 = np.random.randn(100)
y1 = 2 * x1 + np.random.rand(100)

# Conjunto 2
x2 = np.random.randn(100)
y2 = 2 * x2 + np.random.rand(100) + 1  # Deslocamento vertical para contraste

# Gráfico de dispersão comparativo
plt.scatter(x1, y1, color='blue', alpha=0.6, label='Grupo A')
plt.scatter(x2, y2, color='orange', alpha=0.6, label='Grupo B')

plt.title("Comparação entre dois grupos")
plt.xlabel("Variável X")
plt.ylabel("Variável Y")
plt.legend()
plt.show()
```
